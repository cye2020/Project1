```python
import pandas as pd

# 주요 데이터 불러오기
games = pd.read_csv("games.csv")
details = pd.read_csv("games_details.csv")
teams = pd.read_csv("teams.csv")
ranking = pd.read_csv("ranking.csv")

# 상위 5행 확인
print(details.head())
print(details.isnull().sum())
```

```python
int_cols = ['FGM',
       'FGA', 'FG3M', 'FG3A', 'FTM', 'FTA',
       'OREB', 'DREB', 'REB', 'AST', 'STL', 'BLK', 'TO', 'PF', 'PTS']

# 선택한 컬럼만 int로 변환
details[int_cols] = details[int_cols].astype('Int64') 
```

```python
details['AST/TO ratio']=(details['AST']/details['TO'].replace(0,1)).round(2)
```

```python
details['POSS']=(details['FGA']+0.44*details['FTA']-details['OREB']+details['TO']).round(2)
```

```python
import pandas as pd
import numpy as np

def convert_minute(val):
    if pd.isna(val):          # NaN 값 처리
        return 0
    if isinstance(val, str) and ':' in val:   # "MM:SS" 형식
        m, s = val.split(':')
        return int(float(m)) + int(s)/60
    else:
        return float(val)     # 이미 숫자인 경우
```

```python
details['MIN_float']=details['MIN'].apply(convert_minute).round(2)
#분석을 위해 분초를 소수로 바꾸기 
```
